## matplotlib.pyplot.subplot

把一个绘图区域（可以理解成画布）分成多个小区域，用来绘制多个子图。

>matplotlib.pyplot.subplot(*args, **kwargs)[source](https://matplotlib.org/_modules/matplotlib/pyplot.html#subplot)

> subplot(nrows, ncols, index, **kwargs)

- nrows和ncols表示将画布分成（nrows*ncols）个小区域，每个小区域可以单独绘制图形；
- index表示将图绘制在第index个子区域,索引从1到nrows*ncols，以行主顺序递增。

Example

```python
"""
=================
Multiple subplots
=================

Simple demo with multiple subplots.
"""
import numpy as np
import matplotlib.pyplot as plt


x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 2, 2)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 2, 3)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('xxxx')

plt.subplot(2, 2, 4)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```
